```python
# drop rows with no screen specified
data.dropna(subset=['Screen'], inplace=True)
```

```python
# convert currency to TND
data["Final Price"] = data["Final Price"] * 3.1
```

```python
data.head()
```

```python
# strategy 1
# LOW price => <1500 TND
# Medium => < 3000 and >= 1500
# the rest => High
def price_strategy(row):
    if row["Final Price"] < 1500:
        return "Low"
    elif 1500 <= row["Final Price"] < 3000:
        return "Medium"
    else:
        return "High"

data["price"] = data.apply(price_strategy, axis=1)
```

```python
data.describe(include="all")
```

```python
data.head()
```

```python
# strategy2 
# a gaming/design laptop is one which follows the following criteria
# brand: MSI, Razer, Deep gaming
# OR CPU includes "AMD"
# or ram >= 16 and has GPU and , CPU includes "Core i5" or "Core i7" or "Core i9", Storage >= 512

def gaming_strategy(row):
    if row["Brand"] in ["MSI", "Razer", "Deep Gaming"]:
        return True
    if "AMD" in row["CPU"]:
        return True
    if int(row["RAM"]) >= 16 and ("Core i5" in row["CPU"] or "Core i7" in row["CPU"] or "Core i9" in row["CPU"]) and str(row["GPU"]) != "nan" and int(row["Storage"] >= 512):
        return True
    return False
data["usage Gaming and Design"] = data.apply(gaming_strategy, axis=1)
```

```python
data.head()
```

```python
# strategy2 
# an engineering laptop is one which follows the following criteria
# same as gaming, but needs to be SSD Storage Type
# can be without GPU

def engineering_strategy(row):
    if row["Brand"] in ["MSI", "Razer", "Deep Gaming"] and row["Storage type"] == "SSD":
        return True
    if "AMD" in row["CPU"] and row["Storage type"] == "SSD":
        return True
    if int(row["RAM"]) >= 16 and ("Core i5" in row["CPU"] or "Core i7" in row["CPU"] or "Core i9" in row["CPU"]) and int(row["Storage"] >= 512) and row["Storage type"] == "SSD":
        return True
    return False
data["usage Engineering"] = data.apply(engineering_strategy, axis=1)
```

```python
# strategy3
# desk work laptops are all laptops
def desk_work_strategy(row):
    return True

data["usage desk work"] = data.apply(desk_work_strategy, axis=1)
```
